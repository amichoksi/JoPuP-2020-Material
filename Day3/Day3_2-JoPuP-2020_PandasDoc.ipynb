```python
#find duplicate rows based on particular columns
#print(df.duplicated(['customerID']))
# Select all duplicate rows based on one column
duplicateRowsDF = df[df.duplicated(['customerID'])]
print("Duplicate Rows based on a single column are:\n", duplicateRowsDF,"\n")
```

```python
#take average of a column
#From the churn dataframe, what is the average monthly
#charge paid by a customer for the services he/she has 
#signed up for?
#for taking mean, it is required to have that column 
#as number
#if it is not a number, we should convert it to number
#let's remove NA-not available first
df.dropna()

#convert column from string to numeric
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
#By setting errors=’coerce’, transform the non-numeric 
#values into NaN.
print(df['MonthlyCharges'].mean())
```

```python
#In the churn dataframe, under the variable Dependents how many records have “1@#" ?
print(( df[df['Dependents']=="1@#"]==True).count())
```
